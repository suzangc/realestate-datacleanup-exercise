#### Ejercicio 01. ¿Cuál es la casa mas cara en el todo el dataset? (★☆☆)

Imprime la direccion y el precio de la casa seleccionada. Por ejemplo:

`The house with address General Street Nº5 is the most expensive and its price is 5000000 USD`

```python

import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
idvaloralto = ds['price'].idxmax()
dircasa = ds.loc[idvaloralto,'address']
print("La casa de la ", dircasa, "tiene el precio más alto y es $", ds['price'].max())
```

#### Ejercicio 02. ¿Cuál es la casa mas barata del dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`The house with address Concrete Street Nº1 is the cheapest and its price is 12000 USD`

```python

import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
idvalormin = ds['price'].idxmin()
dircasa = ds.loc[idvalormin,'address']
print("La casa de la ", dircasa, "tiene el precio más bajo y es $", ds['price'].min())
```

#### Ejercicio 03. ¿Cuál es la casa mas grande y la mas pequeña del dataset? (★☆☆)

Imprime la dirección y el area de las casas seleccionadas. Por ejemplo:

`The bigger house is located on Yukka Street Nº10 and its surface is 5000 meters`

`The smaller house is located on County Road 1 N and its surface is 200 meters`

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
idvalorx = ds['surface'].idxmax()
dircasax = ds.loc[idvalorx,'address']
idvalorn = ds['surface'].idxmin()
dircasan = ds.loc[idvalorn,'address']

print("La casa más grande está ubicada en ", dircasax, "y mide", ds['surface'].max(), "metros")
print("La casa más pequeña está ubicada en ", dircasan, "y mide", ds['surface'].min(), "metros")
```

#### Ejercicio 04. ¿Cuantas poblaciones (columna level5) contiene el dataset? (★☆☆)

Imprime el nombre de las poblaciones separadas por coma. Por ejemplo:

`> print(populations)`

`population1, population2, population3,...`

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds['level5'] = ds['level5'].apply(lambda x: x.strip()) #intentando quitar los espacios ants y despues de cada valor
cadena = ds['level5'].to_string(index=False, header=False)
poblaciones = ','.join(cadena.split('\n'))
print("Poblaciones: ", poblaciones)
```
